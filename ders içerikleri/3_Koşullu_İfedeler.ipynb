# Koşullu İfadeler

```python
a = -10     # integer
b = 0       # integer
c = 55.5    # float
d = True    # boolean
f = False   # boolean
```

```python
print("IF bloku öncesi")

if True:
    print("if bloku çalıştı.")

if a > b:
    print("A sayısı b sayısından büyük")
if not (b > c):
    print("B sayısı C sayısından büyük")

print("IF bloku sonrası")
```

```python
if a > 0:
    print("A sayısı sıfırdan büyüktür.")
elif a < 0: # else-if
    print("A sayısı sıfırdan küçüktür.")
else:
    print("A sayısı sıfıra eşittir")
```
