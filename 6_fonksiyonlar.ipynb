## Fonksiyonlara Argüman Göndermek

Argüman, doğru kullanıldığı sürece liste, string, int, float, ..... olabilir.

```python
def selamla(isim):
    print(f"Merhaba {isim}")
```

```python
selamla("Ömer Faruk")
```

### varsayılan argüman

```python
def selamla(isim = "insan"):
    print(f"Merhaba {isim}")
```

```python
selamla()
```

## Fonksiyondan Bilgi Döndürmek

```python
def toplama(num1, num2):
    result = num1 + num2
    return result
    
```

```python
toplama(5, 3)
```

### \*ARGS
**Fonksiyonlara Birden Fazla Argüman Yollamak**

```python
def hesapla(*sayi):
    print(type(sayi))
    print(sum(sayi))
```

```python
hesapla(1,2,3,4)
```

```python
def bilgiGoster(*isim):
    for i in isim:
        print(i)
```

```python
bilgiGoster("Ömer","Faruk","Aydın")
```

## **kwargs

```python
def fonksion(**argüman):
    print(type(argüman))
    print(argüman)
```

```python
fonksion(sayi1 = 5, sayı2= 20, sayi3 = 0)
```

## *args ve **kwargs birlikte kullanımı

```python
def fonksiyon(*sayi, **plaka):
    print(sayi)
    print(plaka)
```

```python
fonksiyon(3,4,5, ankara = 6, istanbul = 34)
```

```python
def fonksiyon(num1, num2, *sayi, **plaka):
    print(num1)
    print(num2)
    print(sayi)
    print(plaka)
```

```python
fonksiyon(2, 5,3,5)
```

```python
def fonksiyon(num1, num2, *sayi, plaka):
    print(num1)
    print(num2)
    print(sayi)
    print(plaka)
```

```python
fonksiyon(2, 5,3,5)
```

 ### Kendini Çağıran Fonksiyonlar

 **Recursion Function**

```python
def tri_recursion(k):
    if k > 0:
        result = k + tri_recursion(k - 1)
        print(result)
    else:
        result = 0

    
    return result
```

```python
tri_recursion(5)
```
