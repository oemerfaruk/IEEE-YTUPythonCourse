```python
import sqlite3
```

```python
# 1. Veritabanına bağlan (Yoksa oluşturur)
conn = sqlite3.connect("akademi.db")
cursor = conn.cursor()
```

```python
# 2. Tablo oluşturma
cursor.execute("""CREATE TABLE IF NOT EXISTS student (
               id INTEGER PRIMARY KEY,
               name TEXT,
               surname TEXT,
               department TEXT
               )
""")
```

```python
# 3. Veri ekleme
cursor.execute("""
INSERT INTO student (name, surname, department) VALUES ('Ömer Faruk', 'Aydın', 'Fizik')
""")
```

```python
cursor.execute("""CREATE TABLE IF NOT EXISTS department (
               id INTEGER PRIMARY KEY,
               faculty TEXT,
               department TEXT
               )
""")
```

```python
cursor.execute("""
INSERT INTO department(faculty, department) VALUES ('MF', 'Makine Mühendisliği') 
""")
```

```python
# 4. Değişiklikleri kaydet ve kapat
conn.commit()
conn.close()
```
