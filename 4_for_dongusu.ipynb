## For Döngüleri

```python
numbers = [1,2,3,4,5]
newNumbers = list()
isimler = ["Ömer Faruk", "Osman", "Batuhan", "Mevlüt"]

plakalar = {"Ankara": 6, "Bursa": 16, "İstanbul": 34, "Trabzon": 61}

text = "Hello World!"


a = 0
```

```python
for i in numbers:
    newNumbers.append(pow(i,2))

print("1. For döngüsü bitti.")


for i in newNumbers:
    print(i)

print("2. For döngüsü bitti.")
```

```python
for i, k in plakalar.items():
    print(f"{i} Şehrinin plakası {k}'dır")
```

```python
tekSayilar = list()

for i in range(0,100):
    if i % 2 == 1:
        tekSayilar.append(i)
```

```python
number = int(input("Bir sayı girin: "))
toplam = 0

for i in range(0,number+1):
    toplam = toplam + i

print(toplam)
```

## break - continue

```python
for i in range(0,50):
    print(i)
    if i > 20:
        break
    print(".")
```

```python
for i in range(0,10):
    print("Sayı:")
    if i == 5:
        continue
    
    print(i)
    
```
